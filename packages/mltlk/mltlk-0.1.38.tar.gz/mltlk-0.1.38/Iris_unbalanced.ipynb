## Imports

```python
import mltlk
print(mltlk.__version__)
from mltlk import *
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
```

## Load data

```python
session = load_data("data/iris_unbalanced.csv", 
                    shuffle_data=True, 
                    seed=42,
)
```

#### Show data stats

```python
data_stats(session)
```

## Resample data

```python
split_data(session, test_size=0.15, seed=42)
```

```python
set_resample(session, methods=[
    {"sampler": "u", "max_samples": 40},
    {"sampler": "s", "min_samples": 15},
], seed=42)
```

## Evaluate model

```python
evaluate_model(MLPClassifier(max_iter=1000, random_state=42), 
               session, 
               reload=True,
               mode="split",
               confusionmatrix=True,
)
```
