## Imports

```python
import mltlk
print(mltlk.__version__)
from mltlk import *
from sklearn.neural_network import MLPClassifier
```

## Load data
Load data and scale inputs using standard scaler.

```python
session = load_data("data/spiral.csv", 
                    preprocess="scale",
                    shuffle_data=True,
)
```

#### Show data stats

```python
data_stats(session)
```

## Evaluate model using cross-validation

```python
evaluate_model(MLPClassifier(max_iter=2000, random_state=42), 
               session, 
               reload=False,
               mode="CV-10",
               categories=True,
)
```

## Build final model and predict example
Build final model using all data and predict an unknown example.

```python
build_model(MLPClassifier(max_iter=2000, random_state=42), session)
```

```python
predict([-0.3, 0.3], session)
```

## Plot decision borders

```python
plot_decision_borders(session, cmap="coolwarm", markersize=30, eps=1)
```

## Plot data

```python
plot_data(session, horizontal=False)
```
